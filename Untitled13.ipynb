```python
import zipfile
import os
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms, models
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import gradio as gr

device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
from google.colab import files
uploaded = files.upload()

ZIP_NAME = list(uploaded.keys())[0]

with zipfile.ZipFile(ZIP_NAME, 'r') as zip_ref:
    zip_ref.extractall("dataset")

print("ZIP extraído.")
```

```python
ALLOWED = ["Scarlett Johansson", "Natalie Portman"] # Using available celebrity names

os.makedirs("dataset_two", exist_ok=True)

for person in ALLOWED:
    src = f"dataset/Celebrity Faces Dataset/{person}" # Corrected path
    dst = f"dataset_two/{person}"
    if os.path.exists(src):
        os.system(f"cp -r '{src}' '{dst}'")
    else:
        print(f"Advertencia: La carpeta '{src}' no se encontró. No se copiará.")

print("Dataset filtrado listo.")
```

```python
transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406], [0.229,0.224,0.225])
])
```

```python
dataset = datasets.ImageFolder("dataset_two", transform=transform)
dataset.classes
```

```python
train_size = int(0.7 * len(dataset))
val_size = int(0.2 * len(dataset))
test_size = len(dataset) - train_size - val_size

train_ds, val_ds, test_ds = random_split(dataset, [train_size, val_size, test_size])

train_loader = DataLoader(train_ds, batch_size=16, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=16)
test_loader = DataLoader(test_ds, batch_size=16)
```

```python
model = models.vgg16(weights=models.VGG16_Weights.IMAGENET1K_V1)

for param in model.features.parameters():
    param.requires_grad = False  # congelar convolucionales

model.classifier[6] = nn.Linear(4096, 2)  # 2 clases

model = model.to(device)
model
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.classifier[6].parameters(), lr=0.0001)

EPOCHS = 3

for epoch in range(EPOCHS):
    model.train()
    running_loss = 0

    for imgs, labels in train_loader:
        imgs, labels = imgs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(imgs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Época {epoch+1}/{EPOCHS} - Loss: {running_loss/len(train_loader):.4f}")
```

```python
model.eval()
y_true = []
y_pred = []

with torch.no_grad():
    for imgs, labels in test_loader:
        imgs = imgs.to(device)
        outputs = model(imgs)
        preds = outputs.argmax(1).cpu()

        y_true += labels.tolist()
        y_pred += preds.tolist()

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(cm, display_labels=dataset.classes)
disp.plot(cmap="Blues")
plt.show()
```

```python
CLASSES = dataset.classes

def predict_image(img):
    img = img.resize((224,224))
    img = transform(img).unsqueeze(0).to(device)

    model.eval()
    with torch.no_grad():
        p = model(img).argmax(1).item()

    return f"Predicción: {CLASSES[p]}"

interface = gr.Interface(
    fn=predict_image,
    inputs=gr.Image(type="pil"),
    outputs="text",
    title="Clasificador de Celebridades",
    description="Modelo VGG16 Preentrenado para Scarlett Johansson y Natalie Portman"
)

interface.launch()
```
